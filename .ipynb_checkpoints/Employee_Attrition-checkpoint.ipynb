# This Analysis is for Predicting the Emplooyee Is leave the company or Not.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("files/watson_healthcare_modified.csv")
df.head()
```

```python
attrition_count = pd.DataFrame(df.Attrition.value_counts())
attrition_count
```

So this dataset represents that most of them are like to stay in the company

```python
plt.pie(attrition_count['Attrition'],labels=["Yes","No"],explode=(0.2,0))
```

```python
sns.countplot(data=df,x="Attrition")
```

```python
#Drop the unwanted columns
df = df.drop(['EmployeeCount','EmployeeID'],axis='columns')
```

```python
df.head()
```

```python
attrition_dummies = pd.get_dummies(df.Attrition)
attrition_dummies.head()
```

```python
df = pd.concat([df,attrition_dummies],axis=1)
```

```python
df.head()
```

```python
df.drop(['Attrition','No'],axis=1, inplace=True)
```

```python
df.head()
```

Now i just removed the NO and Attrition features , since the Yes feature says even the employee leave the company or not.

```python
sns.barplot(data=df,x='Gender',y='Yes')
```

```python
sns.barplot(data=df,x='Department',y='Yes')
```

```python
sns.barplot(x='BusinessTravel', y='Yes', data=df)
```

```python
df.drop (['Age', 'JobLevel'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
for column in df.columns:
    if df[column].dtype==np.number:
        continue
    else:
        df[column]=LabelEncoder().fit_transform(df[column])
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X=df.drop('Yes',axis=1)
Y = df.Yes
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
model2 = SVC()
```

```python
model2.fit(X_train,Y_train)
```

```python
model2.score(X_test,Y_test)
```

```python
model3 = SVC(kernel='linear')
```

```python
model3.fit(X_train,Y_train)
```

```python
model3.score(X_test,Y_test)
```
